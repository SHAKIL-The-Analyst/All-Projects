# Exploratory Data Analysis

```python
df.columns
```

### size

```python
size_order = df['Size'].value_counts().index
ax=sns.countplot(x='Size',data=df, palette='Set1',order=size_order, saturation=1)
plt.ylabel("Count")
for bars in ax.containers:
    ax.bar_label(bars)
```

#### Note: From above Graph you can see that most of the people buys M-Size 

# Group By

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Size',y='Quantity', data=S_Qty, palette ='viridis', saturation = 1)
```

##### Note: From above Graph you can see that most of the Qty buys M-Size in the sales 

### Courier Status 

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Courier Status', hue='Status', palette ='Set1', saturation = 1)
```

##### Note: From above Graph the majority of the orders are shipped through the courier.

```python
#histogram 
sns.histplot(df['Size'], kde=False)
```

```python
df['Category'] = df['Category'].astype('str')
plt.figure(figsize=(10, 5))
sns.histplot(df['Category'])
plt.show()
```

##### Note: From above Graph you can see that most of the  buyers buy T-shirt 

```python
# Checking B2B Data  by using pie chart 
B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.title('B2B')
#plt.axis('equal')
plt.show()
```

##### Note : From above chart  we can see that maximum i.e. 99.3% of buyers are retailers and 0.7% are B2B buyers

```python
#  Prepare data for pie chart
a1 = df['Fulfilment'].value_counts()
plt.pie(a1, labels=a1.index, autopct='%1.1f%%')
plt.show()
```

##### Note: From above chart you can see that most of the  Fulfilment  are  amazon 

```python

# Plot the scatter plot
sns.scatterplot(x='Category', y='Size', data=df)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state= df['ship-state'].value_counts(ascending=False).nlargest(10).index
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state)], x='ship-state', palette='Set1', saturation = 1)
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=90)
plt.show()
```

#### Note: From above Graph you can see that most of the  buyers are Maharashtra state

### Conclusion

#### The data analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfills orders through Amazon, experiences high demand for T-shirts, and sees M-Size as the preferred choice among buyers.
