```python
# Assuming 'reservation_status_date' is a string column in your DataFrame
# Convert it to datetime with the correct format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')

# Extract month from the reservation_status_date
df['month'] = df['reservation_status_date'].dt.month

# Plotting
sns.countplot(x='month', hue='is_canceled', data=df, palette='viridis', saturation=1)
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Cancellation Status', labels=['Not Cancelled', 'Cancelled'])
```

```python
sns.barplot(x= 'month', y= 'adr', data=df[df['is_canceled']==1].groupby('month')['adr'].sum().reset_index(),  palette= 'Set1', saturation=1)
plt.ylabel('Average Daily Rate(ADR)')
plt.xlabel('Month')
plt.title('ADR for each month')
```
