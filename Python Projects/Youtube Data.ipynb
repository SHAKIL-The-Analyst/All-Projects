```python
df['Rank']=df['Rank'].str[0:-2]
```

```python
df['Rank']=df['Rank'].str.replace(',', '').astype('int')
df
```

```python
df.columns
```

```python
df['Video Uploads'] = df['Video Uploads'].fillna(0).astype(int)
```

```python
df['Grade']=df['Grade'].str.replace('\xa0 ', 'B')
```

```python
df.dtypes
```

```python
df['Grade']= df['Grade'].map({'A++ ':5, 'A+ ': 4, 'A ': 3, 'A- ': 2, 'B+ ': 1})
```

```python
df['Avg_views']= df['Video views']/df['Video Uploads']
df
```
