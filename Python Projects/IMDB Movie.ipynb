# 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating_category(rating):
    if rating >= 7.0:
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'

# Apply the rating function to create a new column
df['Rating_Category'] = df['Rating'].apply(rating_category)
df
```
